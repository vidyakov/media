# Третье задание
Загрузите цветное изображение из файла img.png. 
Подсчитайте яркость этого изображения и сохраните в файл out_img.png. 
Результирующее изображение должно быть одноканальным. 
Для подсчета яркости используйте формулу Y = 0.2126 \cdot 
R + 0.7152 \cdot G + 0.0722 \cdot B Y=0.2126⋅R+0.7152⋅G+0.0722⋅B , 
не забудьте сначала перевести изображение в вещественные числа 
(функция img_as_float), а затем в целые числа (функция img_as_ubyte).

```python
from skimage import img_as_ubyte
from skimage.color import rgb2gray
```

```python
img = img_as_float(imread('https://stepik.org/media/attachments/lesson/58181/tiger-color.png'))
r, g, b = [img[:, :, i] for i in range(3)]
```

```python
gray = 0.2126 * r + 0.7152 * g + 0.0722 * b
imshow(gray)
```

```python
imsave('out_img.png', img_as_ubyte(gray))
```
