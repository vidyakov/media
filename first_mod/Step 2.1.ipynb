# Четвертое задание
Загрузите изображение из файла img.png. Изображение состоит из рамки сплошного цвета и внутренней части изображения. Цвет рамки можно узнать, посмотрев на левый верхний пиксель. Рамка может иметь разную ширину со всех четырех сторон. Определите размеры рамки и выведите эти размеры через пробел. Размеры рамки выводите в следующем порядке: левый, верхний, правый, нижний.

```python
img = imread('https://stepik.org/media/attachments/lesson/58180/img.png')
imshow(img)
```

```python
height, weight = img.shape
border_color = img[0, 0]
border_color, height, weight
```

```python
borders_sizes = []
```

```python
# top
def get_height():
    size = 0
    for row in range(height):
        for col in range(weight):
            if img[row, col] != border_color:
                return size
        size += 1
```

```python
borders_sizes.append(get_height())
```

```python
# bottom
def get_bottom():
    size = 0
    for row in reversed(range(height)):
        for col in range(weight):
            if img[row, col] != border_color:
                return size
        size += 1
```

```python
borders_sizes.append(get_bottom())
```

```python
# left
def get_left():
    size = 0
    for col in range(weight):
        for row in range(height):
            if img[row, col] != border_color:
                return size
        size += 1
```

```python
borders_sizes.append(get_left())
```

```python
# right
def get_right():
    size = 0
    for col in reversed(range(weight)):
        for row in range(height):
            print(row, col)
            if img[row, col] != border_color:
                return size
        size += 1
```
